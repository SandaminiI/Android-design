```python
from datasets import load_dataset

dataset = load_dataset("mwitiderrick/arXiv", split="train[:5000]")  # small subset for Colab
dataset = dataset.train_test_split(test_size=0.02, seed=42)
train_ds = dataset["train"]
val_ds = dataset["test"]

print(train_ds[0])
```

```python
def add_dummy_summaries(batch):
    batch["summary"] = ["This is a dummy summary." for _ in batch["text"]]
    return batch

train_ds = train_ds.map(add_dummy_summaries, batched=True)
val_ds = val_ds.map(add_dummy_summaries, batched=True)
```

```python
from transformers import AutoTokenizer

model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)

max_input_length = 1024
max_target_length = 150

def preprocess_function(batch):
    inputs = [i[:20000] for i in batch["text"]]  # use the correct column name
    # If your target summaries are in another field, replace "summary" below accordingly
    model_inputs = tokenizer(
        inputs,
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )

    # Use text_target instead of as_target_tokenizer()
    model_inputs["labels"] = tokenizer(
        batch["summary"],
        max_length=max_target_length,
        truncation=True,
        padding="max_length"
    )["input_ids"]

    return model_inputs

tokenized_train = train_ds.map(preprocess_function, batched=True)
tokenized_val   = val_ds.map(preprocess_function, batched=True)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="/content/bart-arxiv", tokenizer=tokenizer, device=0)

text = """Paste introduction or body text of any scientific paper here..."""
summary = summarizer(text, max_length=200, min_length=60, do_sample=False)
print("📄 Summary:\n", summary[0]['summary_text'])
```
