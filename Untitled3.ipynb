```python
# --- Imports ---
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# Make plots look nice
plt.style.use('seaborn-v0_8')
```

```python
class GridWorld:
    def __init__(self):
        self.height = 8
        self.width = 8
        self.grid = np.zeros((self.height, self.width)) - 1  # default reward -1
        self.current_location = (4, np.random.randint(0, 5))
        self.bomb_location = (1, 3)
        self.gold_location = (0, 3)
        self.terminal_states = [self.bomb_location, self.gold_location]
        self.grid[self.bomb_location] = -10
        self.grid[self.gold_location] = 10
        self.actions = ['UP', 'DOWN', 'LEFT', 'RIGHT']

    def get_available_actions(self):
        return self.actions

    def agent_on_map(self):
        grid = np.zeros((self.height, self.width))
        grid[self.current_location] = 1
        return grid

    def get_reward(self, new_location):
        return self.grid[new_location]

    def make_step(self, action):
        last_location = self.current_location
        if action == 'UP':
            if last_location[0] > 0:
                self.current_location = (last_location[0]-1, last_location[1])
        elif action == 'DOWN':
            if last_location[0] < self.height-1:
                self.current_location = (last_location[0]+1, last_location[1])
        elif action == 'LEFT':
            if last_location[1] > 0:
                self.current_location = (last_location[0], last_location[1]-1)
        elif action == 'RIGHT':
            if last_location[1] < self.width-1:
                self.current_location = (last_location[0], last_location[1]+1)

        reward = self.get_reward(self.current_location)
        return reward

    def check_state(self):
        if self.current_location in self.terminal_states:
            return 'TERMINAL'
```

```python
class QNetwork(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(QNetwork, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(state_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, action_dim)
        )

    def forward(self, x):
        return self.layers(x)
```

```python
class DQNAgent:
    def __init__(self, environment, epsilon=0.1, lr=0.001, gamma=0.9):
        self.env = environment
        self.epsilon = epsilon
        self.gamma = gamma
        self.actions = environment.get_available_actions()
        self.action_dim = len(self.actions)
        self.state_dim = 2  # (row, column)
        self.model = QNetwork(self.state_dim, self.action_dim)
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.loss_fn = nn.MSELoss()

    def get_state_tensor(self, state):
        # Normalize by grid size for stability
        norm_state = [state[0]/self.env.height, state[1]/self.env.width]
        return torch.FloatTensor(norm_state)

    def choose_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.actions)  # explore
        else:
            state_tensor = self.get_state_tensor(state)
            with torch.no_grad():
                q_values = self.model(state_tensor)
            action_index = torch.argmax(q_values).item()
            return self.actions[action_index]

    def learn(self, old_state, action, reward, new_state, done):
        old_state_tensor = self.get_state_tensor(old_state)
        new_state_tensor = self.get_state_tensor(new_state)

        q_values = self.model(old_state_tensor)
        next_q_values = self.model(new_state_tensor).detach()

        action_index = self.actions.index(action)
        target = reward + self.gamma * torch.max(next_q_values) * (1 - int(done))

        loss = self.loss_fn(q_values[action_index], target)

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
```

```python
def play_dqn(environment, agent, episodes=500, max_steps=500):
    rewards_per_episode = []
    for ep in range(episodes):
        cumulative_reward = 0
        done = False
        step = 0
        while not done and step < max_steps:
            old_state = environment.current_location
            action = agent.choose_action(old_state)
            reward = environment.make_step(action)
            new_state = environment.current_location

            done = environment.check_state() == 'TERMINAL'
            agent.learn(old_state, action, reward, new_state, done)

            cumulative_reward += reward
            step += 1

            if done:
                environment.__init__()  # reset environment
        rewards_per_episode.append(cumulative_reward)
    return rewards_per_episode
```

```python
epsilons = [0.1, 0.5, 0.9]
all_rewards = {}

for eps in epsilons:
    env = GridWorld()
    agent = DQNAgent(env, epsilon=eps)
    rewards = play_dqn(env, agent, episodes=500)
    all_rewards[eps] = rewards

plt.figure(figsize=(8,5))
for eps, rewards in all_rewards.items():
    plt.plot(rewards, label=f"Îµ = {eps}")
plt.title("GridWorld Deep Q-Learning - Epsilon Comparison")
plt.xlabel("Episode")
plt.ylabel("Total Reward per Episode")
plt.legend()
plt.show()
```

```python
def smooth(y, box_pts=10):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.figure(figsize=(8,5))
for eps, rewards in all_rewards.items():
    plt.plot(smooth(rewards), label=f"Îµ = {eps} (smoothed)")
plt.title("Smoothed Learning Curves for DQN Agent")
plt.xlabel("Episode")
plt.ylabel("Smoothed Total Reward")
plt.legend()
plt.show()
```
